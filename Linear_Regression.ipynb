```python
# Importing the necessary libraries
import matplotlib.pyplot as plt  # For plotting graphs
import numpy as np  # For numerical operations
import pandas as pd  # For handling data in dataframes
```

```python
# Reading the CSV file and storing the data in a DataFrame
df = pd.read_csv('/content/Placements.csv')
```

```python
# Displaying the first 5 rows
df.head()
```

```python
# Scatter plot of CGPA vs Package (LPA)
plt.scatter(df['CGPA'], df['Package (LPA)'])
plt.xlabel('CGPA')  # X-axis label
plt.ylabel('Package (LPA)')  # Y-axis label
```

```python
# Defining feature (X) and target (y) variables
X = df.iloc[:, 0:1]  # Feature: CGPA
y = df.iloc[:, -1]   # Target: Package (LPA)
```

```python
# Displaying the target variable (y) to check its values
y
```

```python
# Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split  # Importing the train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  # 20% data for testing
```

```python
# Importing the LinearRegression class
from sklearn.linear_model import LinearRegression

# Creating an instance of the LinearRegression model
regressor = LinearRegression()
```

```python
# Fitting the linear regression model to the training data
regressor.fit(X_train, y_train)
```

```python
# Displaying the test feature set (X_test) to check its values
X_test
```

```python
# Displaying the test target variable (y_test) to check its values
y_test
```

```python
# Predicting the target value for the second test sample
# Reshaping the input to 2D array format (1, 1) for the model
regressor.predict(X_test.iloc[1].values.reshape(1, 1))
```

```python
# Scatter plot of CGPA vs Package (LPA) with the regression line
plt.scatter(df['CGPA'], df['Package (LPA)'])  # Scatter plot of actual data
plt.plot(X_train, regressor.predict(X_train), color='red')  # Regression line based on training data
plt.xlabel('CGPA')  # X-axis label
plt.ylabel('Package (LPA)')  # Y-axis label
```

```python
# Retrieving and printing the coefficients of the linear regression model
m = regressor.coef_
print(m)  # Displaying the slope of the regression line
```

```python
# Retrieving and printing the intercept of the linear regression model
b = regressor.intercept_
print(b)  # Displaying the y-intercept of the regression line
```
